```python
from transformers import pipeline

# Create a summarization pipeline
# summarizer = pipeline("summarization")

# Function to split a long sentence into chunks and summarize each chunk
def summarize_long_sentence(long_sentence, chunk_size=500, overlap=50):
    chunks = [long_sentence[i:i+chunk_size] for i in range(0, len(long_sentence), chunk_size - overlap)]
    
    summaries = []
    for chunk in chunks:
        summary = summarizer(chunk, max_length=50, min_length=40, do_sample=False)
        summaries.append(summary[0]['summary_text'])
    
    return summaries

# Example usage
# input_text = "Your big sentence goes here..."
# summaries = summarize_long_sentence(input_text)

# Print summaries for each chunk
# for i, summary in enumerate(summaries):
#     print(f"Summary for chunk {i + 1}:", summary)
```

```python
from pytube import Search
from youtube_transcript_api import YouTubeTranscriptApi, NoTranscriptFound

def fetch_manual_captions(video_url, languages=['en', 'a.en']):
    try:
        manual_captions = YouTubeTranscriptApi.get_transcript(video_url, languages=languages)
        return [segment['text'] for segment in manual_captions]
    except NoTranscriptFound:
        # Return an empty list when no transcripts are found
        return []
    except Exception as e:
        # Print a message or log the error if needed
        return []

def save_captions_to_file(captions, output_file='op.txt'):
    with open(output_file, 'a', encoding='utf-8') as opf:
        opf.write('\n'.join(captions) + '\n')

def main():
    query = 'g20 english'
    si = Search(query)

    top_5_results = si.results[:5]
    ans=""
    
    for result in top_5_results:
        video_url = result.video_id
        try:
            # Fetch both generated and manually created captions
            all_captions = YouTubeTranscriptApi.get_transcript(video_url)

            # Skip videos without any available transcripts
            if not any(all_captions):
                print(f"Skipping video {video_url} - No transcripts available.")
                continue

            # Fetch manually created English transcripts
            manual_captions = fetch_manual_captions(video_url)
            
            save_captions_to_file(manual_captions)
            print(result.title)
            texti = ''.join(manual_captions)
            print(len(texti))
            if len(texti)<5000:
                result_summary=summarize(texti)
            else: 
                result_summary=summarize_long_sentence(texti)
            

            
            # print(f"Manually provided captions available: {bool(manual_captions)}")
            ans+="".join(str(result_summary))
            # print(count_sentences(texti))
            # print(result_summary)
        except NoTranscriptFound:
            print(f"No transcript found for video {video_url}. Skipping.")
    # re1=count_sentences(ans) // 2
    # final_summary=generate_summary(ans,re1)
    # print(re1)
    print(ans)
    audio_text="".join(str(ans))
    text_to_speech(audio_text)

if __name__ == "__main__":
    main()
```
